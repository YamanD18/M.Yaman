```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df = pd.read_csv("data kecelakaan.csv")

X = df[['Jumlah Pelanggaran Lalu lintas']]
y = df['Jumlah Kecelakaan']

model = LinearRegression()
model.fit(X,y)

df['Prediksi jumlah kecelakaan'] = model.predict(X)

plt.scatter(X,y,color='blue')
plt.plot(X, df['Prediksi jumlah kecelakaan'],color='red')
plt.xlabel("Jumlah Pelanggaran Lalu lintas")
plt.ylabel("Jumlah Kecelakaan")
plt.title("Regresi linear pelanggaran vs kecelakaan")
plt.grid(True)
plt.show

print("koefisien (dampak pelenggaran ke kecelakaan):", model.coef_[0])
print("intercept (titik potong):", model.intercept_)
```
